## 2. Stufenfunktionen
Die Stufenfunktion dient dazu, zu unterscheiden, wann das Neuron feuert und wann nicht. 
Bei einem einzelnen Perzeptron wurde als Stufenfunktion die Heaviside-Funktion verwendet.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def entscheidung( summe ):
    if summe >=0 :  return 1
    else:        return 0
x =  np.linspace(-1, 1, 100, endpoint=True)
y = [ entscheidung(i) for i in x ]
# Erzeugen des Graphen, mit einer blauen Stufe und Bezeichnung step
plt.step(x, y, '-b', label='step')
plt.grid(True)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.xlabel('Summe')
plt.ylabel('Ergebnis')
plt.title('Stufenfunktion')
plt.legend(loc='lower right')
plt.show()
```

Für den Gradientenabstieg muss diese Stufenfunktion allerdings differenzierbar sein, so dass die Heaviside-Funktion nicht geeignet ist. Man verwendet deshalb die so ganannte Sigmoid-Funktion, die ein ähnliches Verhalten wie letztere hat.
<span class="math">$$ \sigma (x) = \frac 1 {1 + e ^{-c\cdot x} }, c > 0$$ </span> 
Je größer der Faktor c ist, desto mehr nähert sich die Funktion der Heaviside-Funktion an.

```python
x =  np.linspace(-3, 3, 100, endpoint=True)
y1 = [ 1/(1 + np.exp(-i)) for i in x ]
y2 = [ 1/(1 + np.exp(-2*i)) for i in x ]
y3 = [ 1/(1 + np.exp(-10*i)) for i in x ]
fig, ax1 = plt.subplots()
ax1.plot(x, y1, label="c=1")
ax1.plot(x,y2, label="c=2")
ax1.plot(x,y3, label="c=10")
ax1.set_title('Sigmoidfunktionen im Vergleich')
ax1.grid(True)
legend = ax1.legend(loc='best', fancybox=True)
plt.show()
```

Im weiteren werden wir die einfachste Variante mit c = 1 benutzen. 

Die Ableitung der Sigmoid-Funktion lässt sich leicht mit der Produkt- und der Kettenregel berechnen und dann in diese Form bringen (NACHRECHNEN!!):
<span class="math">$$ \sigma (x)' =  \sigma (x) \cdot ( 1- \sigma(x))$$ </span> 
Eine weitere Stufenfunktion, die oft benutzt wird, ist tanh. Auch hier ist die Ableitung in eine einfache Form zu bringen:
 
<span class="math">$$\tanh x=\frac{\sinh x}{\cosh x}=\frac{e^x -e^{-x}}{e^x+e^{-x}}=\frac{e^{2x}-1}{e^{2x}+1}=1-\frac{2}{e^{2x }+1}$$    
    
    
$$\frac{\mathrm{d}}{\mathrm{d}x} \tanh x = 1-\tanh^2 x $$</span> 
Diese Funktion bildet allerdings auf das Intervall [-1;1] ab.

```python
x =  np.linspace(-3, 3, 100, endpoint=True)
y = [np.tanh(i) for i in x ]
fig, ax1 = plt.subplots()
ax1.plot(x, y, label="tanh")
ax1.set_title('tanh')
ax1.grid(True)
legend = ax1.legend(loc='best', fancybox=True)
plt.show()
```
