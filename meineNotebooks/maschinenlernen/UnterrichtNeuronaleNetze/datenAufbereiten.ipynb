```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13]#.values
y = dataset.iloc[:, 13]#.values
X = pd.get_dummies(X, columns=["Geography"], prefix=["Land_is"] )
X = pd.get_dummies(X, columns=["Gender"], prefix=["G_is"] )
```

# Aufbereitung von Daten
* Datensatz und Quelle: https://medium.com/@pushkarmandot/build-your-first-deep-learning-neural-network-model-using-keras-in-python-a90b5864116d 
* Einlesen der Daten

* Aufteilen in Trainingsdaten und Testdaten
* Skalieren der Merkmale:
 * MinMaxScaler: Mappt die Merkmale zwischen den angegebenen Bereich, hier 0 und 1 
 * StandardScaler: Mappt die Merkmale so, dass der Mittelwert 0 ist und die Standardabweichung 1

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
from sklearn import preprocessing # import StandardScaler
#sc = StandardScaler() #als Alternative
sc =preprocessing.StandardScaler()#MinMaxScaler(feature_range=(0, 1))
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_test[0:3]
```

## Netzwerk bauen
* 13 Inputs, 2 Hidden Schichten mit je 6 Neuronen und einen Output
* Aktivierung mit relu

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense( 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 13))
# Adding the second hidden layer
classifier.add(Dense( 6, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(1, kernel_initializer= 'uniform', activation = 'sigmoid'))
```

```python
# Compiling Neural Network
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting our model 
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
